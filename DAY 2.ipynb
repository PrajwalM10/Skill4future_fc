```python
# OOPS on BS
class Bank:
    def __init__(self, balance):
        #self.amount = amount
        self.balance = balance
    # I want to deposit
    def deposit(self, amount):
        # balance + amount
        self.balance = self.balance + amount
        # if amount is neg add if else statement
    # I want to withdraw
    def withdraw(self, amount):
       # pass
        # balance - amount
        if amount > self.balance:
            print('Insufficient fund!')
        else:
            self.balance = self.balance - amount
    def balance(self):
        return self.balance
```

```python
acc = Bank(0)
```
