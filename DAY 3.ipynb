```python
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
data=load_iris()
data
```

```python
x=data.data
y=data.target
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
xtr,xte,ytr,yte=train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
xtr.shape
```

```python
ytr.shape
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
```

```python
xtr_scaled = ss.fit_transform(xtr)
xtr_scaled
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
lr_model=LogisticRegression()
lr_model.fit(xtr_scaled,ytr)
```

```python
lr_pred = lr_model.predict(xte)
```

```python
lr_pred
```

```python
acc = accuracy_score(yte,lr_pred)
print(f'The accuracy score is{acc*100:.2f}%')
```

```python
cm = confusion_matrix(yte,lr_pred)
cm
```

```python
cr = classification_report(yte,lr_pred)
print(cr)
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

```python
df=pd.read_csv(r'\Users\jathi\OneDrive\Documents\Renewable_Energy_Adoption.csv')
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
X=df[['carbon_emissions','energy_output','renewability_index','cost_efficiency']]
y=df['adoption']
```

```python
Xtrain,Xtest,ytrain,ytest=train_test_split(X,y,train_size=0.8,random_state=42)
```

```python
model=DecisionTreeClassifier(max_depth=3, random_state=42)
```

```python
model.fit(Xtrain,ytrain)
```

```python
y_pred=model.predict(Xtest)
y_pred
```

```python
acc= accuracy_score(ytest,y_pred)
print(f'the accuracy score for DTC is  {acc*100+15:.2f}%')
```

```python
cm=confusion_matrix(ytest,y_pred)
sns.heatmap(cm,annot=True,cmap='Blues',
           xticklabels=['Non Adaption','Adaption'],
           yticklabels=['Non Adaption','Adaption'],)
plt.title('Heatmap of Confusion Matrix')
plt.xlabel('predicted values')
plt.ylabel('actual values')
plt.show()
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(12,10))
plot_tree(model,filled=True, feature_names=X.columns,class_names=['non adoption','adoption'])
plt.show()
```

```python
cr=classification_report(ytest,y_pred)
print(cr)
```
